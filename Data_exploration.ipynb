## Load Packages and Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import csv 
import os
```

## Load Dataset

```python
#Load in dataset services 2019-2022
serv19 = pd.read_csv("data/Train Services 2019-2025/services-2019.csv")
serv20 = pd.read_csv("data/Train Services 2019-2025/services-2020.csv")
serv21 = pd.read_csv("data/Train Services 2019-2025/services-2021.csv")
serv22 = pd.read_csv("data/Train Services 2019-2025/services-2022.csv")
```

```python
#Load in dataset services 2023-2025
serv23 = pd.read_csv("data/Train Services 2019-2025/services-2023.csv")
serv24 = pd.read_csv("data/Train Services 2019-2025/services-2024.csv")
serv25 = pd.read_csv("data/Train Services 2019-2025/services-2025.csv")
```

```python
#Load in dataset disruptions 2019-2022
dis19 = pd.read_csv("data/Train Disruptions 2019-2025/disruptions-2019.csv")
dis20 = pd.read_csv("data/Train Disruptions 2019-2025/disruptions-2020.csv")
dis21 = pd.read_csv("data/Train Disruptions 2019-2025/disruptions-2021.csv")
dis22 = pd.read_csv("data/Train Disruptions 2019-2025/disruptions-2022.csv")
```

```python
#Load in dataset disruptions 2023-2025
dis23 = pd.read_csv("data/Train Disruptions 2019-2025/disruptions-2023.csv")
dis24 = pd.read_csv("data/Train Disruptions 2019-2025/disruptions-2024.csv")
dis25 = pd.read_csv("data/Train Disruptions 2019-2025/disruptions-2025.csv")
```

```python
#Load in dataset weather 2019-2025
weather = pd.read_csv( "data/Weather 2019-2025/Weather 2019-2025.txt", sep= ',', skiprows= 46, header= 0 )
```

```python
#Concatenate dataset services as one

train_services = pd.concat([serv19, serv20, serv21, serv22, serv23, serv24, serv25])
serv_col_names = train_services.columns
print( f" The column names are: {serv_col_names}" )

# rows, columns = train_services.shape
# print( f"train_services has Rows: {rows}, Columns: {columns}" ) #Rows: 38571, Columns: 14
```
