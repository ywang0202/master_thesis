```python
#check for Missing Value 
#code from 12:59 https://www.youtube.com/watch?v=MoM6mighOJM&t=778s 

total_serv = train_services.isnull().sum().sort_values(ascending=False)
percent_serv = (train_services.isnull().sum()/train_services.isnull().count().sort_values(ascending=False))
missvalue_services = pd.concat([total_serv,percent_serv], axis= 1, keys=["Total", "Percent"])
missvalue_services
```

```python
#Check for any duplicates in the dataset
train_services.duplicated().sum()
```

```python
#Removing the rows with missing values
serv_cleaned = train_services.dropna()

print("Dataset services after removing rows with missing values:")
print(serv_cleaned)

#Old [38571 rows x 14 columns]
#New [38446 rows x 14 columns]
#removed 125 rows
```

```python
#Univariate Analysis
#right-skewed with extreme outliers

# Check for outliers 
print(train_services['duration_minutes'].describe())
print(f"\n99th percentile: {train_services['duration_minutes'].quantile(0.99):.0f} minutes")
print(f"Max: {train_services['duration_minutes'].max():.0f} minutes")

for p in [90, 95, 99, 99.9]:
    value = train_services['duration_minutes'].quantile(p/100)
    count = (train_services['duration_minutes'] > value).sum()
    print(f"Top {100-p}% (> {value:.0f} min): {count} disruptions")
```
