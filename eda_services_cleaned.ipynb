```python
#read the csv file
for i in ["2019", "2020", "2021", "2022", "2023", "2024", "2025"]:
    print(f"loading {i}")
    data = pd.read_csv(f"I:/Datasets Thesis/Train Services 2019-2025/services-{i}.csv")
    
    print(f"Data shape before cleaning: {data.shape}")
    
    print(f"Starting Cleaning {i}")
    
    # Convert time
    data["Service:Date"] = pd.to_datetime(
    data["Service:Date"].astype(str).str[:10],
    format="%Y-%m-%d",
    errors="coerce")
    
    # Drop all rows where Service:Type is NOT in the list
    data = data[data["Service:Type"].isin(["Sprinter", "Stoptrein", "Intercity", "Intercity direct", "Extra trein", "Sneltrein"])]
    print(f"Data shape after filtering Service:Type: {data.shape}")

   # Keep only NS company
    data = data[data["Service:Company"] == "NS"]
    print(f"Data shape after filtering Service:Company: {data.shape}")
    
    # Keep only rows where Stop:Station name is NOT in the non-Dutch stations list. Remove all non-Dutch stations 
    data = data[data["Stop:Station name"].isin(dutch_stations)]
    print(f"Data shape after dropping non-Dutch stations: {data.shape}")
    
    print(f"Data shape after cleaning: {data.shape}")
    print(f"Finished Cleaning {i}")
    
    # # # When done export to csv
    print(f"Exporting Clean {i}")
    data.to_csv(f'I:/Datasets Thesis/Train Services 2019-2025/clean/services_clean_{i}.csv', index=False)
```
