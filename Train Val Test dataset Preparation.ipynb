```python
import pandas as pd
```

```python
data = pd.read_csv("combined_data_clean.csv")

# Define chunk size (bigger for millions of rows)
chunk_size = 1_000_000

# Split DataFrame into chunks
chunks = [data.iloc[i:i + chunk_size] for i in range(0, len(data), chunk_size)]

# Display chunk sizes (avoid printing full chunk contents)
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}: {chunk.shape}")
```

```python
# Combine chunks back into DataFrames
train_df = pd.concat(chunks[:50], ignore_index=True)
val_df   = pd.concat(chunks[50:60], ignore_index=True)
test_df  = pd.concat(chunks[60:], ignore_index=True)

# Verify sizes
print("\nFinal split sizes:")
print(f"Train: {train_df.shape}")
print(f"Validation: {val_df.shape}")
print(f"Test: {test_df.shape}")
```

```python
# Save the split datasets to CSV files
train_df.to_csv("train_data.csv", index=False)
val_df.to_csv("val_data.csv", index=False)
test_df.to_csv("test_data.csv", index=False)
```
